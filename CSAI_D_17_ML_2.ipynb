#Data Cleaning

```python
df = pd.read_csv('/content/sample_data/Building_Permits.csv')
df.head()
```

Check dataset info

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

Detect missing values

```python
df.isnull().sum()
```

Handle missing values

```python
df_drop = df.dropna()
df_drop
```

```python
df['Estimated Cost'].fillna(df['Estimated Cost'].mean(),inplace = True)
```

```python
df['Completed Date'].fillna(df['Completed Date'].mode(),inplace = True)
```

```python
df = df.fillna(df.mode().iloc[0])
df
```

Remove duplicate records

```python
df.drop_duplicates(inplace = True)
```

Handle incorrect datatypes

```python
df['Completed Date'] = pd.to_datetime(df['Completed Date'])
```

```python
df['Record ID'] = df['Record ID'].astype(float)
```

Handle Outliers

```python
Q1 = df['Revised Cost'].quantile(0.25)
Q3 = df['Revised Cost'].quantile(0.75)
IQR = Q3 - Q1
```

```python
df = df[(df['Revised Cost'] >= Q1 - 1.5*IQR) & (df['Revised Cost'] <= Q3 + 1.5*IQR)]
```

Standardize

```python
df['Revised Cost'] = (df['Revised Cost'] - df['Revised Cost'].mean()) / df['Revised Cost'].std()
```

Encode categorical data

```python
df['Street Number Suffix'] = df['Street Number Suffix'].map({'St':0,'Av':1})
```

```python
df = pd.get_dummies(df, columns = ['Proposed Construction Type Description'])
```

Save dataset

```python
df.to_csv('/content/sample_data/Building_Permits.csv',index = False)
```
