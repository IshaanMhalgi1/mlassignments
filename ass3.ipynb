```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df
```

Scatter plot

```python
plt.scatter(df['total_rooms'],df['median_house_value'],label='Total Rooms')
plt.scatter(df['population'],df['median_house_value'],label='Population')
plt.title("Total Rooms and Population vs Median House Value")
plt.legend()
plt.grid()
plt.ylabel("Median house value")
plt.show()
```

Bar plot

```python
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']
plt.bar(df['latitude'],df['population'],color = bar_colors)
plt.title("Latitude vs Population")
plt.legend()
plt.xlabel("Latitude")
plt.ylabel("Population")
plt.show()
```

Box Plot

```python
plt.boxplot(df['median_house_value'])
plt.grid()
plt.title('median house value Box Plot')
```

Pie Chart

```python
x = np.array(np.random.randint(0,200,5))
y = np.array(['USA','China','India','UK','France'])

plt.title('GDP Distribution')
plt.pie(x,labels=y)
plt.legend()
```

Line plot

```python
plt.plot(df['total_rooms'],label='Total Rooms')
plt.plot(df['total_bedrooms'],label='Total Bedrooms')
plt.title('Rooms and Bedrooms')
plt.legend()
plt.grid()
plt.xlabel('Total members in households')
plt.show()
```
